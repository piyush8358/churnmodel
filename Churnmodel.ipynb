```python
x=dataset[['RowNumber', 'CustomerId', 'Surname', 'CreditScore', 'Geography',
       'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard',
       'IsActiveMember', 'EstimatedSalary']]
```

```python
Gender=pd.get_dummies(Gender,drop_first=True)
```

```python
HasCrCard=dataset['HasCrCard']
```

```python
IsActiveMember=dataset['IsActiveMember']
```

```python
x.drop( ["RowNumber", "CustomerId", "Surname" , "Geography" , "Gender" , "NumOfProducts","HasCrCard", "IsActiveMember" ], axis=1, inplace=True)
```

```python
x
```

```python
x=pd.concat([x,Gender,Geography,NumOfProducts,IsActiveMember,HasCrCard],axis=1)
```

```python
from keras.models import Sequential
```

```python
model=Sequential()
```

```python
model.get_config()
```

```python
from keras.layers import Dense
```

```python
# layer 1
model.add(
Dense(
units=6 ,
input_dim=13,
bias_initializer="zeros",
kernel_initializer="zeros",
activation="relu"
)
)
```

```python
model.get_config()
```

```python
model.summary()
```

```python
# Layer 2
model.add(
Dense(
units=7 ,
bias_initializer="zeros",
kernel_initializer="zeros",
activation="relu"
)
)
```

```python
model.get_config()
```

```python
model.summary()
```

```python
#Layer 3 
model.add(
Dense(
units=5,
bias_initializer="zeros",
kernel_initializer="zeros",
activation="relu"
)
)
```

```python
model.get_config()
```

```python
model.summary()
```

```python
# layer 4 : output layer
model.add(
Dense(
units=1,
bias_initializer="zeros",
kernel_initializer="zeros",
activation="sigmoid"
)
)
```

```python
model.summary()
model.get_config()


```

```python
model.compile(loss="binary_crossentropy")
```

```python
model.fit(x, y, epochs=20)
```
